# ***Simple Calculator***


```python
def add(a, b):
  return a + b

def subtract(a, b):
  return a - b

def multiply(a, b):
  return a * b

def divide(a, b):
  return a / b
```

```python
ans_addition = add(3, 4)
ans_subtraction = subtract(3, 4)
ans_multiplication = multiply(3, 4)
ans_division = divide(3,4)
```

```python
print(ans_addition)
print(ans_subtraction)
print(ans_multiplication)
print(ans_division)
```
